```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country
```

```python
new_df =  pd.merge(df,df_country,on='Country Code',how='left')
```

```python
new_df.head()
```

```python
new_df.Country.value_counts()
```

```python
country_names = new_df.Country.value_counts().index
```

```python
country_val = new_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.f%%')
plt.show()
```

```python
ratings = new_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. From 4.5 to 4.9 --> Excellent Rating
2. From 4.0 to 4.4 --> very good Rating
3. From 3.5 to 3.9 --> good
4. From 2.5 to 3.4 --> average
5. From 1.8 to 2.4 --> Poor

```python
ratings.head()
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color')
```

## Observation:-
1. Not Rated count is very high 
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
